```python
#Since, we don't need customerID, We drop the column
data = data.drop('customerID',axis=1)
data.head()
```

```python
#Converting Object column to Numerical Column, which is actually holds Numerical data
data['TotalCharges'] = pd.to_numeric(data.TotalCharges, errors='coerce')
data.isna().sum()
```

```python
#Removing tenure equal to 0
data.drop(labels=data[data['tenure'] == 0].index, axis=0, inplace=True)
data[data['tenure'] == 0].index
```

```python
#Categorizing Senior Citizen column for encoding
data["SeniorCitizen"]= data["SeniorCitizen"].map({0: "No", 1: "Yes"})
data.head()
```

```python
# Label Encoding is used since tree-based models are not affected by ordinal encoding
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
le = LabelEncoder()
Scaler = StandardScaler()
```

```python
for col in data.columns:
    if data[col].dtypes=='object':
        data[col] = le.fit_transform(data[col])
```

```python
data.head(5)
```

```python
# Dropping PhoneService and gender to reduce multicollinearity and low predictive impact
x = data.drop(columns = ['Churn','PhoneService','gender'])
x
```

```python
y= data['Churn'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.25,random_state=69)
```

```python
x_train
```

```python
x_test
```

```python
num_cols = ["tenure", 'MonthlyCharges', 'TotalCharges']
```

```python
x_train[num_cols] = Scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = Scaler.transform(x_test[num_cols])
```

# **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import *
```

```python
model_RFC = RandomForestClassifier( n_estimators=100,        # number of trees
    random_state=10)
```

```python
model_RFC.fit(x_train,y_train)
```

```python
# Make predictions
prediction_test = model_RFC.predict(x_test)
```

```python
# Print accuracy
print(accuracy_score(y_test, prediction_test))
```

```python
confusion_matrix(y_test,prediction_test)
```

```python
print(classification_report(y_test,prediction_test))
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_LR = LogisticRegression()
```

```python
model_LR.fit(x_train,y_train)
```

```python
# Make predictions
prediction_test = model_LR.predict(x_test)
```

```python
confusion_matrix(y_test,prediction_test)
```
